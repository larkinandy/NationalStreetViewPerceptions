```python
import forestplot as fp
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
```

```python
PARENT_FOLDER = 'insert absolute filepath to where files are stored'
labels = ['SafeCrime','SafeWalk','Beauty','Nature','Relaxing']
urbanData = []
for label in labels:
    urbanData.append(ps.read_csv(PARENT_FOLDER + "urbanForest" + label + ".csv"))
```

```python
# the code in this cell is based on examples provided by the forestplot homepage at https://pypi.org/project/forestplot/
for index in range(len(labels)):
    tempPlot = fp.forestplot(urbanData[index],  # the dataframe with results data
                             estimate="score",  # col containing estimated effect size 
                             ll="score", hl="score",  # columns containing conf. int. lower and higher limits
                             varlabel="val",  # column containing variable label
                             capitalize="capitalize",  # Capitalize labels
                             groupvar="cat",  # Add variable groupings 
                             group_order=["all", "age","education","ethnicity","gender","race","covid"],
                             sort=False,  # sort in ascending order (sorts within group if group is specified)               
                             color_alt_rows=True,  # Gray alternate rows
                             ci_report=False,
                             figsize=(6,12)
                            )
    tempPlot.set_xlim(-14,14)
    # Save just the portion _inside_ the second axis's boundaries
    plt.savefig(PARENT_FOLDER + "ploturban" + labels[index] + ".svg", bbox_inches="tight")
```
